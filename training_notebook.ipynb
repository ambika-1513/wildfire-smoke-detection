```python
import dagshub
dagshub.init(repo_owner='ambika-patil', repo_name='wildfire-smoke-detection', mlflow=True)

import mlflow
import yaml

with open("params.yaml") as f:
    params = yaml.safe_load(f)

with mlflow.start_run(run_name="yolov8s_exp_v0"):
    # Log parameters
    mlflow.log_params(params)

    # Log metrics
    mlflow.log_metric("mAP50", 0.917)
    mlflow.log_metric("mAP50-95", 0.529)
    mlflow.log_metric("precision", 0.825)
    mlflow.log_metric("recall", 0.895)

    # Log model
    mlflow.log_artifact("models/model.pt")
```
